```python
#Cargo el primer dataset, con los datos
puestosPrivados=pd.read_csv("puestos_privados_mensual_por_clae6.csv")
puestosPrivados
```

```python
#Cargo el segundo dataset con el detalle y descripción
clavesPuestos=pd.read_csv("clae_agg.csv")
clavesPuestos
```

```python
#Realizo la unión de ambos, por medio del código clae6
tablaUnion=pd.merge(left=clavesPuestos,right=puestosPrivados,left_on="clae6",right_on="clae6")
tablaUnion
```

```python
#Separo la fecha, en día, mes y año
tablaGlobal=tablaUnion
tablaGlobal["Fecha"]=tablaGlobal["fecha"].astype(str)
tablaGlobal["Año"]=tablaGlobal["Fecha"].str[0:4]
tablaGlobal["Mes"]=tablaGlobal["Fecha"].str[5:7]
tablaGlobal["Día"]=tablaGlobal["Fecha"].str[8:10]
tablaGlobal
```

```python
#Gráfico que muestra la cantidad de empleos totales creados en cada Año
puestosAnio=tablaGlobal.groupby("Año")["puestos"].sum()
puestosAnio.plot(kind="line",grid=True)
plt.title("Cantidad de empleos creados por año en millones (1e7)")
plt.xlabel("Año")
plt.ylabel("Cantidad de Empleos")
plt.show()
```

```python
#Gráfico que indica la cantidad de empleos totales creados por cada Sector
puestoSector=tablaGlobal.groupby("letra_desc")["puestos"].sum()
puestoSector.plot(kind="barh",grid=True)
plt.title("Cantidad Total de Empleos generados por Sector en millones")
plt.xlabel("Cantidad de Empleos")
plt.ylabel("Sector")
plt.show()
```
